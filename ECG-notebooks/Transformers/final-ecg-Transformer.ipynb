```python
###########################################
# 0. Imports
###########################################
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy.signal import spectrogram
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, f1_score
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight

import tensorflow as tf
import tensorflow.keras.backend as K
from tensorflow.keras.layers import (Input, Dense, LayerNormalization, Dropout,
                                     MultiHeadAttention, GlobalAveragePooling1D,
                                     Add)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping

!pip install imbalanced-learn --quiet
from imblearn.over_sampling import RandomOverSampler

###########################################
# 1. Load and Preprocess ECG Data
###########################################
data = loadmat('ECGData.mat')
ecg_data = data['ECGData']
ecg_data_item = ecg_data[0, 0]

Signals = ecg_data_item['Data']  # shape: (162, 65536)
Labels_array = ecg_data_item['Labels']

# Convert labels to list of strings
Labels = [Labels_array[i, 0][0] for i in range(Labels_array.shape[0])]

# Map labels: NSR -> 0 (Normal), ARR/CHF -> 1 (Abnormal), else -> 2 (unused)
Labels_mapped = []
for label in Labels:
    if label == 'NSR':
        Labels_mapped.append(0)
    elif label in ['ARR', 'CHF']:
        Labels_mapped.append(1)
    else:
        Labels_mapped.append(2)
Labels_mapped = np.array(Labels_mapped)

# (Optional) Filter out label=2 if it exists
keep_idx = np.where(Labels_mapped < 2)[0]
Signals = Signals[keep_idx]
Labels_mapped = Labels_mapped[keep_idx]

# Standardize signals
Signals_normalized = []
for signal in Signals:
    signal = signal.astype(np.float32)
    mean_val = np.mean(signal)
    std_val = np.std(signal) if np.std(signal) > 1e-6 else 1e-6
    signal = (signal - mean_val) / std_val
    Signals_normalized.append(signal)
Signals_normalized = np.array(Signals_normalized, dtype=np.float32)

print("Signals shape:", Signals_normalized.shape)
print("Labels shape:", Labels_mapped.shape)

unique, counts = np.unique(Labels_mapped, return_counts=True)
print("Label distribution after filtering (0=Normal,1=Abnormal):")
for u,c in zip(unique, counts):
    print(f"  Label {u}: {c} samples")

###########################################
# 2. Segment the Data (10s windows + 50% overlap)
###########################################
def segment_signals(signals, labels, sample_time_size=10, fs=128):
    """
    Segments ECG signals into windows of length sample_time_size (sec).
    Uses 50% overlap by default -> step_size=window_size//2
    """
    window_size = fs * sample_time_size  # e.g., 10s * 128 = 1280
    step_size = window_size // 2         # 50% overlap

    segs, seg_labels = [], []
    for i in range(len(signals)):
        sig = signals[i]
        lbl = labels[i]
        if len(sig) < window_size:
            continue  # skip if the signal is too short

        for start in range(0, len(sig) - window_size + 1, step_size):
            seg_data = sig[start : start + window_size]
            segs.append(seg_data)
            seg_labels.append(lbl)

    return np.array(segs, dtype=np.float32), np.array(seg_labels, dtype=np.int32)

X_segments, Y_segments = segment_signals(Signals_normalized, Labels_mapped, sample_time_size=10, fs=128)
print("Segmented X shape:", X_segments.shape)  
print("Segmented Y shape:", Y_segments.shape)

###########################################
# 3. Feature Extraction (Spectrogram-based)
###########################################
fs = 128

def compute_instantaneous_frequency_spectrogram(signal, fs=128):
    f, t, Sxx = spectrogram(signal, fs=fs, nperseg=64, noverlap=32)
    Sxx = np.abs(Sxx)
    Sxx_norm = Sxx / np.sum(Sxx, axis=0, keepdims=True)
    inst_freq = np.sum(f[:, np.newaxis] * Sxx_norm, axis=0)
    return inst_freq, t

def compute_spectral_entropy(signal, fs=128):
    f, t, Sxx = spectrogram(signal, fs=fs, nperseg=64, noverlap=32)
    Sxx = np.abs(Sxx)
    psd = Sxx / np.sum(Sxx, axis=0, keepdims=True)
    spectral_entropy = -np.sum(psd * np.log(psd + 1e-12), axis=0)
    spectral_entropy /= np.log(psd.shape[0])
    return spectral_entropy, t

def extract_features_dataset(signals, fs=128):
    """
    For each 1D signal, compute [inst_freq, spectral_entropy].
    Output => list of shape (2, time_steps).
    """
    features = []
    for signal in signals:
        inst_freq, _ = compute_instantaneous_frequency_spectrogram(signal, fs)
        spec_entropy, _ = compute_spectral_entropy(signal, fs)

        min_len = min(len(inst_freq), len(spec_entropy))
        inst_freq = inst_freq[:min_len]
        spec_entropy = spec_entropy[:min_len]
        
        feat_matrix = np.vstack((inst_freq, spec_entropy))  # shape: (2, min_len)
        features.append(feat_matrix)
    return features

X_features = extract_features_dataset(X_segments, fs=128)

# Zero-pad to MAX length
max_length = max(f.shape[1] for f in X_features)
X_features_padded = []
for feat in X_features:
    pad_width = max_length - feat.shape[1]
    if pad_width > 0:
        feat_padded = np.pad(feat, ((0,0),(0,pad_width)), mode='constant')
    else:
        feat_padded = feat
    X_features_padded.append(feat_padded)

X_final = np.array([f.T for f in X_features_padded], dtype=np.float32)  # (num_samples, max_time, 2)
print("X_final shape:", X_final.shape)

###########################################
# 4. Train-Test Split
###########################################
# Convert labels to 0/1
le = LabelEncoder()
Y_segments_encoded = le.fit_transform(Y_segments)  # 0->0, 1->1 typically

X_train, X_test, y_train, y_test = train_test_split(
    X_final, 
    Y_segments_encoded,
    test_size=0.2, 
    stratify=Y_segments_encoded, 
    random_state=42
)

X_train, X_valid, y_train, y_valid = train_test_split(
    X_train, 
    y_train, 
    test_size=0.2, 
    stratify=y_train, 
    random_state=42
)

print("Train:", X_train.shape, y_train.shape)
print("Valid:", X_valid.shape, y_valid.shape)
print("Test: ",  X_test.shape,  y_test.shape)

###########################################
# 5. Define a Larger Transformer + Focal Loss
###########################################
def focal_loss(gamma=2.0, alpha=0.5):
    """
    Focal Loss for binary classification.
    FL = -alpha*(1-pt)^gamma * log(pt)
    """
    def focal_loss_fixed(y_true, y_pred):
        y_pred = K.clip(y_pred, 1e-7, 1-1e-7)
        pt = tf.where(tf.equal(y_true, 1), y_pred, 1 - y_pred)
        return -K.mean(alpha * K.pow((1-pt), gamma) * tf.math.log(pt))
    return focal_loss_fixed

def positional_encoding(sequence_length, d_model):
    """Generates positional encoding for transformer input."""
    pos = np.arange(sequence_length)[:, np.newaxis]  # (seq_len, 1)
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))

    pos_encoding = np.zeros((sequence_length, d_model))
    pos_encoding[:, 0::2] = np.sin(pos * div_term)
    pos_encoding[:, 1::2] = np.cos(pos * div_term)
    return tf.convert_to_tensor(pos_encoding, dtype=tf.float32)

def transformer_encoder(inputs, head_size=64, num_heads=4, ff_dim=256, dropout=0.1):
    attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=head_size)(inputs, inputs)
    attn_output = Dropout(dropout)(attn_output)
    x = Add()([attn_output, inputs])
    x = LayerNormalization(epsilon=1e-6)(x)

    ffn = Dense(ff_dim, activation='relu')(x)
    ffn = Dropout(dropout)(ffn)
    ffn = Dense(inputs.shape[-1])(ffn)  # project back to d_model
    x = Add()([ffn, x])
    x = LayerNormalization(epsilon=1e-6)(x)
    return x

sequence_length = X_train.shape[1]
d_input = X_train.shape[2]
d_model = 128               # Larger embedding
num_heads = 4
ff_dim = 256               # Larger feed-forward dimension
dropout = 0.1
num_transformer_blocks = 3  # 3 blocks instead of 2

inputs = Input(shape=(sequence_length, d_input))
x = Dense(d_model)(inputs)

pos_encoding = positional_encoding(sequence_length, d_model)
x = x + pos_encoding

for _ in range(num_transformer_blocks):
    x = transformer_encoder(x, head_size=64, num_heads=num_heads,
                            ff_dim=ff_dim, dropout=dropout)

x = GlobalAveragePooling1D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
outputs = Dense(1, activation='sigmoid')(x)

model = Model(inputs, outputs)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss=focal_loss(gamma=2.0, alpha=0.5),  # use focal loss
    metrics=['accuracy']
)

model.summary()

###########################################
# 6. Partial Oversampling w/ Ratio=0.75
###########################################
desired_ratio = 0.75  # minority is 75% the size of majority
from collections import Counter

counter_train = Counter(y_train)
print("Original train distribution:", counter_train)

major_class = 0 if counter_train[0] > counter_train[1] else 1
minor_class = 1 if major_class==0 else 0

major_count = counter_train[major_class]
minor_count = counter_train[minor_class]

target_minority = int(desired_ratio * major_count)
print("target_minority:", target_minority)

X_train_2d = X_train.reshape(X_train.shape[0], -1)

if target_minority > minor_count:
    sampling_strategy = {
        minor_class: target_minority,
        major_class: major_count
    }
else:
    sampling_strategy = 'minority'

print("Sampling strategy:", sampling_strategy)
ros = RandomOverSampler(sampling_strategy=sampling_strategy, random_state=42)

X_resampled, y_resampled = ros.fit_resample(X_train_2d, y_train)
X_resampled = X_resampled.reshape(-1, sequence_length, d_input)

res_counter = Counter(y_resampled)
print("Resampled distribution:", res_counter)

###########################################
# 7. Training
###########################################
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_resampled, y_resampled,
    validation_data=(X_valid, y_valid),
    epochs=50,
    batch_size=64,
    callbacks=[early_stopping],
    verbose=1
)

plt.figure()
plt.plot(history.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'], label='Val loss')
plt.plot(history.history['accuracy'], label='Train acc')
plt.plot(history.history['val_accuracy'], label='Val acc')
plt.legend()
plt.title("Training Curves - Focal Loss & Larger Transformer")
plt.show()

###########################################
# 8. Threshold Tuning
###########################################
def find_best_threshold_for_f1(y_true, y_probs, step=0.01):
    best_thresh, best_f1 = 0, -1
    for t in np.arange(0,1+step,step):
        preds = (y_probs>t).astype(int)
        score = f1_score(y_true, preds)
        if score>best_f1:
            best_f1=score
            best_thresh=t
    return best_thresh, best_f1

y_valid_probs = model.predict(X_valid).ravel()
best_thresh, best_f1 = find_best_threshold_for_f1(y_valid, y_valid_probs)
print(f"Best threshold on validation: {best_thresh:.2f}, F1={best_f1:.4f}")

###########################################
# 9. Evaluate on Test
###########################################
y_test_probs = model.predict(X_test).ravel()
y_test_preds = (y_test_probs>best_thresh).astype(int)

cm_test = confusion_matrix(y_test, y_test_preds)
report_test = classification_report(y_test, y_test_preds, digits=4)
print("\nConfusion Matrix (Test):")
print(cm_test)
print("\nClassification Report (Test):")
print(report_test)

###########################################
# 10. Evaluate on Training
###########################################
y_train_probs = model.predict(X_train).ravel()
y_train_preds = (y_train_probs>best_thresh).astype(int)
cm_train = confusion_matrix(y_train, y_train_preds)
report_train = classification_report(y_train, y_train_preds, digits=4)

print("\nConfusion Matrix (Train):")
print(cm_train)
print("\nClassification Report (Train):")
print(report_train)
```
